```python
# Kolb 설문조사 예시 코드
def kolb_survey():
    print("Kolb 학습 유형 설문조사에 오신 것을 환영합니다.")
    print("각 질문에 대해 1부터 5까지 점수를 입력해 주세요 (1: 매우 그렇지 않다, 5: 매우 그렇다).\n")

    # 각 항목에 대한 질문
    questions = [
        "1. 구체적인 경험을 통해 배울 때 효과가 좋다고 생각한다.",
        "2. 다른 사람들과 토론하는 과정에서 배운다.",
        "3. 이론을 학습하고 개념을 이해하는 데 중점을 둔다.",
        "4. 실험을 통해 배운다.",
        # 필요한 만큼 질문을 추가
    ]

    # 학습 스타일 점수를 저장할 변수
    ce_score = 0  # 구체적 경험
    ro_score = 0  # 반성적 관찰
    ac_score = 0  # 추상적 개념화
    ae_score = 0  # 활동적 실험

    # 질문을 순차적으로 출력하면서 사용자 입력을 받음
    for i, question in enumerate(questions):
        print(question)
        score = int(input("점수 (1-5): "))

        # 각 질문에 맞는 학습 스타일 점수를 추가 (예시는 임의로 각 학습 스타일에 배정함)
        if i % 4 == 0:
            ce_score += score
        elif i % 4 == 1:
            ro_score += score
        elif i % 4 == 2:
            ac_score += score
        elif i % 4 == 3:
            ae_score += score

    # 결과 출력
    print("\nKolb 학습 스타일 결과:")
    print(f"구체적 경험(CE): {ce_score}")
    print(f"반성적 관찰(RO): {ro_score}")
    print(f"추상적 개념화(AC): {ac_score}")
    print(f"활동적 실험(AE): {ae_score}")

    # 학습 스타일 분석
    if ce_score > ac_score and ro_score > ae_score:
        print("당신의 학습 스타일은 반성적 경험형입니다.")
    elif ce_score > ac_score and ae_score > ro_score:
        print("당신의 학습 스타일은 활동적 경험형입니다.")
    elif ac_score > ce_score and ro_score > ae_score:
        print("당신의 학습 스타일은 반성적 이론형입니다.")
    elif ac_score > ce_score and ae_score > ro_score:
        print("당신의 학습 스타일은 활동적 이론형입니다.")
    else:
        print("균형 잡힌 학습 스타일을 가지고 있습니다.")

# 설문조사 실행
kolb_survey()
```

```python
# Kolb 학습 유형 검사를 위한 함수 정의
def kolb_learning_test(responses):
    # CE(구체적 경험), RO(반성적 관찰), AC(추상적 개념화), AE(능동적 실험)를 위한 변수 초기화
    CE, RO, AC, AE = 0, 0, 0, 0

    # 각 질문에 대한 선택지를 학습 차원으로 매핑
    # responses[i] = [순위1, 순위2, 순위3, 순위4] 에 사용자가 선택한 순위가 들어감
    question_mapping = [
        [0, 1, 2, 3],  # 1번 질문의 선택지를 CE, RO, AC, AE에 매핑
        [0, 1, 2, 3],  # 2번 질문
        [0, 1, 2, 3],  # 3번 질문
        [1, 0, 3, 2],  # 4번 질문의 매핑이 다를 수 있음
        [3, 2, 1, 0],  # 5번 질문의 예시 매핑
        # 나머지 12개 질문에 대한 매핑 추가
    ]

    # 각 응답을 순회하면서 점수 누적
    for i, response in enumerate(responses):
        ranks = question_mapping[i]  # 현재 질문에 대한 매핑 가져오기
        CE += 5 - response[ranks[0]]  # 1순위 = 4점, 4순위 = 1점
        RO += 5 - response[ranks[1]]
        AC += 5 - response[ranks[2]]
        AE += 5 - response[ranks[3]]

    # 각 차원의 평균 계산
    CE_avg = CE / len(responses)
    RO_avg = RO / len(responses)
    AC_avg = AC / len(responses)
    AE_avg = AE / len(responses)

    # 두 축(AC-CE와 AE-RO) 계산
    AC_CE = AC_avg - CE_avg
    AE_RO = AE_avg - RO_avg

    # 학습 유형 결정
    if AC_CE <= 7 and AE_RO <= 6:
        learning_style = '확산형 (Diverging)'
    elif AC_CE >= 8 and AE_RO <= 6:
        learning_style = '동화형 (Assimilating)'
    elif AC_CE >= 8 and AE_RO >= 7:
        learning_style = '수렴형 (Converging)'
    elif AC_CE <= 7 and AE_RO >= 7:
        learning_style = '적용형 (Accommodating)'
    else:
        learning_style = '분류되지 않음'

    # 결과 반환
    return {
        "CE 평균": CE_avg,
        "RO 평균": RO_avg,
        "AC 평균": AC_avg,
        "AE 평균": AE_avg,
        "AC-CE 값": AC_CE,
        "AE-RO 값": AE_RO,
        "학습 유형": learning_style
    }

# 예시 사용
# 각 내부 리스트는 한 질문의 네 가지 선택지에 대한 순위(1순위~4순위)를 나타냄
example_responses = [
    [1, 2, 3, 4],  # 1번 질문 응답
    [2, 1, 4, 3],  # 2번 질문 응답
    [4, 3, 2, 1],  # 3번 질문 응답
    [1, 3, 4, 2],  # 4번 질문 응답
    [3, 4, 1, 2],  # 5번 질문 응답
    # 12개의 질문 모두 추가
]

# 테스트 실행
result = kolb_learning_test(example_responses)
print(result)
```

```python
def kolb_survey():
    print("Kolb 학습 유형 설문조사에 오신 것을 환영합니다.")
    print("각 질문에 대해 4가지 선택지를 1~4순위로 매겨주세요 (1순위가 가장 높은 점수입니다).\n")

    # 각 항목에 대한 질문과 선택지
    # 각 항목에 대한 질문과 선택지
    questions = [
        {
            "question": "1. 나는 학습할 때",
            "choices": [
                "기분에 좌우되는 편이다 (CE4)",
                "주의 깊게 보고 듣는 것을 좋아한다 (RO)",
                "새로운 아이디어에 대해 생각하는 것을 좋아한다 (AC)",
                "직접 해보는 것을 좋아한다 (AE)"
            ]
        },
        {
            "question": "2. 나는 _ 학습이 가장 잘 된다",
            "choices": [
                "내 직감을 믿을 때 (CE)",
                "신중하게 보고 들을 때 (RO)",
                "논리적으로 생각할 때 (AC)",
                "주어진 과제를 끝내기 위해 열심히 노력할 때 (AE)"
            ]
        },
        {
            "question": "3. 나는 배울 때",
            "choices": [
                "감정적인 반응 및 대응이 빠른 편이다 (CE)",
                "조용하고 내성적이다 (RO)",
                "이유를 알아내려는 경향이 있다 (AC)",
                "학습에 책임감을 느낀다 (AE)"
            ]
        },
                {
            "question": "4. 나는",
            "choices": [
                "느낌을 통해 배운다 (CE)",
                "보면서 배운다 (RO)",
                "생각하면서 배운다 (AC)",
                "직접 해보면서 배운다 (AE)"
            ]
        },
                        {
            "question": "5. 나는 학습할 때",
            "choices": [
                "새로운 경험에 대하여 개방적이다 (CE)",
                "문제를 여러 측면에서 고려한다 (RO)",
                "문제를 분석하여 종류별로 분류하는 것을 좋아한다 (AC)",
                "직접 실행해 보는 것을 좋아한다 (AE)"
            ]
        },
                        {
            "question": "6. 학습할 때 나는",
            "choices": [
                "직관적인 사람이다 (CE)",
                "관찰력 있는 사람이다 (RO)",
                "논리적인 사람이다 (AC)",
                "행동적인 사람이다 (AE)"
            ]
        },
                        {
            "question": "7. 나는 _ 가장 학습이 잘 된다",
            "choices": [
                "사람들과 교류하면서 (CE)",
                "관찰할 때 (RO)",
                "합리적인 이론을 통해 (AC)",
                "연습 및 실습의 기회가 많을 때 (AE)"
            ]
        },
                        {
            "question": "8. 나는 학습할 때",
            "choices": [
                "학습에 몰입한다 (CE)",
                "먼저 생각하고 행동하는 편이다 (RO)",
                "아이디어와 이론을 좋아한다 (AC)",
                "나의 작업결과를 확인하는 것을 좋아한다 (AE)"
            ]
        },
                        {
            "question": "9. 나는 _ 가장 잘 학습한다",
            "choices": [
                "느낌에 의존할 때 (CE)",
                "관찰할 때 (RO)",
                "내 생각에 의존할 때 (AC)",
                "스스로 실행해 볼 수 있을 때 (AE)"
            ]
        },
                        {
            "question": "10. 나는 학습하는 동안",
            "choices": [
                "남의 의견을 수용하는 편이다 (CE)",
                "나의 의견을 표현하지 않는 편이다 (RO)",
                "이성적인 사람이다 (AC)",
                "책임감 있는 사람이다 (AE)"
            ]
        },
                        {
            "question": "11. 나는 학습할 때~",
            "choices": [
                "그룹에 참여하여 활동하는 것을 좋아하는 편이다 (CE)",
                "관찰하는 것을 좋아한다 (RO)",
                "신중하게 내용을 평가한다 (AC)",
                "활동적인 것을 좋아한다 (AE)"
            ]
        },
                        {
            "question": "12. 나는 _ 가장 잘 배운다",
            "choices": [
                "수용적이고 개방적인 상황에서 (CE)",
                "주의를 기울일 때 (RO)",
                "아이디어를 분석 할 때 (AC)",
                "실제적이고 실행해 볼 수 있을 때 (AE)"
            ]
        },
        # 더 많은 질문을 여기에 추가할 수 있습니다.
    ]

    # 학습 스타일 점수를 저장할 변수
    ce_score = 0  # 구체적 경험
    ro_score = 0  # 반성적 관찰
    ac_score = 0  # 추상적 개념화
    ae_score = 0  # 활동적 실험

    # 질문을 순차적으로 출력하면서 사용자 입력을 받음
    for q in questions:
        print(q["question"])
        print("각 선택지를 순위에 맞춰 1부터 4까지 입력하세요 (순위는 중복되지 않게 입력해주세요):")
        ranks = []
        entered_ranks = set()  # 입력된 순위를 추적

        for i, choice in enumerate(q["choices"]):
            while True:
                try:
                    rank = int(input(f"{choice}: "))
                    if rank < 1 or rank > 4:
                        print("순위는 1부터 4 사이의 값이어야 합니다.")
                    elif rank in entered_ranks:
                        print("중복된 순위를 입력하셨습니다. 다시 입력해주세요.")
                    else:
                        entered_ranks.add(rank)
                        ranks.append((i, rank))  # 선택지의 인덱스와 순위를 함께 저장
                        break
                except ValueError:
                    print("유효한 숫자를 입력해주세요.")

        # 순위를 점수로 변환 (1순위 = 4점, 2순위 = 3점, 3순위 = 2점, 4순위 = 1점)
        scores = [0, 0, 0, 0]  # CE, RO, AC, AE 점수
        for index, rank in ranks:
            scores[index] = 5 - rank  # 1순위일 경우 4점, 4순위일 경우 1점

        # 점수 누적
        ce_score += scores[0]  # CE 점수
        ro_score += scores[1]  # RO 점수
        ac_score += scores[2]  # AC 점수
        ae_score += scores[3]  # AE 점수

    # 결과 출력
    print("\nKolb 학습 스타일 결과:")
    print(f"구체적 경험(CE): {ce_score}")
    print(f"반성적 관찰(RO): {ro_score}")
    print(f"추상적 개념화(AC): {ac_score}")
    print(f"활동적 실험(AE): {ae_score}")

    # 축 계산
    ac_ce_diff = ac_score - ce_score  # AC-CE 축
    ae_ro_diff = ae_score - ro_score  # AE-RO 축


    # 학습 스타일 분석
    if ac_ce_diff <= 7 and ae_ro_diff <= 6:
        print("당신의 학습 스타일은 확산형 (Diverging)입니다.")
    elif ac_ce_diff >= 8 and ae_ro_diff <= 6:
        print("당신의 학습 스타일은 동화형 (Assimilating)입니다.")
    elif ac_ce_diff >= 8 and ae_ro_diff >= 7:
        print("당신의 학습 스타일은 수렴형 (Converging)입니다.")
    elif ac_ce_diff <= 7 and ae_ro_diff >= 7:
        print("당신의 학습 스타일은 적용형 (Accommodating)입니다.")
    else:
        print("분류되지 않은 학습 스타일입니다.")

# 설문조사 실행
kolb_survey()
```
